```python
fig, axs = plt.subplots(2,4)
fig.set_figheight(10)
fig.set_figwidth(20)
axs[0, 0].scatter(df2['Core 0 load (%)'],df2['Core 0 Temp. (°)'],color='orange')
axs[0, 0].set_title('Core 0')
axs[0, 1].scatter(df2['Core 1 load (%)'],df2['Core 1 Temp. (°)'],color='red')
axs[0, 1].set_title('Core 1')
axs[0, 2].scatter(df2['Core 2 load (%)'],df2['Core 2 Temp. (°)'],color='purple')
axs[0, 2].set_title('Core 2')
axs[0, 3].scatter(df2['Core 3 load (%)'],df2['Core 3 Temp. (°)'],color='green')
axs[0, 3].set_title('Core 3')

axs[1, 0].scatter(df2['Core 4 load (%)'],df2['Core 4 Temp. (°)'],color='blue')
axs[1, 0].set_title('Core 4')
axs[1, 1].scatter(df2['Core 5 load (%)'],df2['Core 5 Temp. (°)'],color='brown')
axs[1, 1].set_title('Core 5')
axs[1, 2].scatter(df2['Core 6 load (%)'],df2['Core 6 Temp. (°)'],color='fuchsia')
axs[1, 2].set_title('Core 6')
axs[1, 3].scatter(df2['Core 7 load (%)'],df2['Core 7 Temp. (°)'],color='skyblue')
axs[1, 3].set_title('Core 7')
for ax in axs.flat:
    ax.set(xlabel='Load', ylabel='Temp')
```
